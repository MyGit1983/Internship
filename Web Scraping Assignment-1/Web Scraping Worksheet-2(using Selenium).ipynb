```python
driver.get('https://www.naukri.com/')
```

```python
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Analyst')
search_loc = driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
search_loc.send_keys('Bangalore')
```

Finding element for job search bar.

```python
search_btn = driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()
```

```python
url='https://www.naukri.com/data-analyst-jobs-in-bangalore?k=data%20analyst&l=bangalore'
```

Give the Url.

```python
driver.get(url)
```

```python
job_titles=[]
job_location=[]
company_name=[]
experience_list=[] 
```

Created 4 empty lists. In these lists the data will be stored while scraping.

```python
titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
```

Now we have all the tags in which there are the job titles.

Now we will extract the text from these tags one by one by looping over thees tags.

```python
for i in titles_tags:
    title=i.text
    job_titles.append(title)
#job_titles[0:10]
```

```python
location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
#location_tags[0:10]
    
```

```python
for i in location_tags:
    job_location.append(i.text)
    
#job_location[0:10]
    
```

```python
company_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
```

```python
for i in company_tags:
    company_name.append(i.text)
#company_name[0:10]
```

```python
experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
#experience_tags[0:10]
```

```python
for i in experience_tags:
    experience_list.append(i.text)
#experience_list[0:10]
```

```python
jobs=pd.DataFrame()
jobs['title']=job_titles
jobs['location']=job_location
jobs['company']=company_name
jobs['experience_required']=experience_list
jobs[0:10]
```

Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to 
scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount %
The attributes which you have to scrape is ticked marked in the below image.
To scrape the data you have to go through following steps:
1. Go to flipkart webpage by url https://www.flipkart.com/
2. Enter “sunglasses” in the search field where “search for products, brands and 
more” is written and click the search icon
3. after that you will reach to a webpage having a lot of sunglasses. From this page 
you can scrap the required data as usual.
4. after scraping data from the first page, go to the “Next” Button at the bottom of 
the page , then click on it
5. Now scrape data from this page as usual
6. repeat this until you get data for 100 sunglasses.

Note that all of the above steps have to be done by coding only and not manually.

```python
import selenium
import pandas as pd
from selenium import webdriver

driver = webdriver.Chrome(r"C:\Users\Ankita\Downloads\chromedriver_win32/chromedriver.exe")
driver.get('https://www.flipkart.com')

search_product = driver.find_element_by_xpath("//input[@class='_3704LK']")
search_product.send_keys('sunglasses')                                             
search_btn = driver.find_element_by_xpath("//button[@type='submit']")
search_btn.submit()

url='https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'

driver.get(url)

brand_list=[]
description_list=[]
price_list=[]
discount_list=[]

brand_tags=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")


for i in brand_tags:
    brand_list.append(i.text)
brand_list

description_tags=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
description_tags
for i in description_tags:
    description_list.append(i.text)
description_list

price_tags=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
price_tags
for i in price_tags:
    price_list.append(i.text)
price_list

discount_tags=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
discount_tags
for i in discount_tags:
    discount_list.append(i.text)
discount_list

data=pd.DataFrame()
data['brand']=brand_list[:35]
data['description']=description_list[:35]
data['price']=price_list[:35]
data['discount']=discount_list[:35]
data
```

```python
import selenium
import pandas as pd
from selenium import webdriver

driver = webdriver.Chrome(r"C:\Users\Ankita\Downloads\chromedriver_win32/chromedriver.exe")
driver.get('https://www.amazon.in/')
search_product = driver.find_element_by_xpath("//input[@id='twotabsearchtextbox']")
search_product.send_keys('Laptop')

search_btn = driver.find_element_by_xpath("//input[@id='nav-search-submit-button']")
search_btn.click()

filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_button:
    if i.text=='Intel Core i7':
        i.click()
        break
        
filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_button:
    if i.text=='Intel Core i9':
        i.click()
        break
        
title_list=[]
rating_list=[]
price_list=[]

titles_tags=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")
#titles_tags[0:10]

for i in titles_tags:
    title_list.append(i.text)
#print(title_list[0:10])

rating_tags=driver.find_elements_by_xpath("//span[@class='a-icon-alt']")


for i in rating_tags:
    rating_list.append(i.text)
    
#print(rating_list[0:10])

price_tags=driver.find_elements_by_xpath("//span[@class='a-price-whole']")


for i in price_tags:
    price_list.append(i.text)
#print(price_list[0:10])

data=pd.DataFrame()
data['title']=title_list
#data['rating']=rating_list
data['price']=price_list
data[0:10]

driver.close
```
